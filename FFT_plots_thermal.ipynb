### How to analyze FFT for the frequency? 

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()
fig = plt.figure(figsize=(32,12))
plt.subplots_adjust(wspace=0.0, hspace=0)
plt.rcParams.update({'font.size': 12})
```

```python

img_V1 = cv.imread('paper_plots/18_real_A.png',0)
img_T1 = cv.imread('paper_plots/18_real_B.png',0)
img_V2 = cv.imread('paper_plots/60_real_A.png',0)
img_T2 = cv.imread('paper_plots/60_real_B.png',0)

def my_fft(img):
    f = np.fft.fft2(img)
    fshift = np.fft.fftshift(f)
    magnitude_spectrum = 20*np.log(np.abs(fshift))
    return magnitude_spectrum

def my_fft_plots(img_V1, img_T1, img_V2, img_T2):

    plt.figure(figsize=(22,10))
    plt.subplot(241),plt.imshow(img_V1, cmap = 'gray')
    plt.title('Input Image'), plt.xticks([]), plt.yticks([])

    plt.subplot(242),plt.imshow(my_fft(img_V1), cmap = 'gray')
    plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])

    plt.subplot(243),plt.imshow(img_T1, cmap = 'gray')
    plt.title('Input Image'), plt.xticks([]), plt.yticks([])

    plt.subplot(244),plt.imshow(my_fft(img_T1), cmap = 'gray')
    plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])

    # SECOND SAMPLES
    plt.subplot(245),plt.imshow(img_V2, cmap = 'gray')
    plt.title('Input Image'), plt.xticks([]), plt.yticks([])

    plt.subplot(246),plt.imshow(my_fft(img_V2), cmap = 'gray')
    plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])

    plt.subplot(247),plt.imshow(img_T2, cmap = 'gray')
    plt.title('Input Image'), plt.xticks([]), plt.yticks([])

    plt.subplot(248),plt.imshow(my_fft(img_T2), cmap = 'gray')
    plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
img_V1 = cv.imread('paper_plots/18_real_A.png',0)
img_T1 = cv.imread('paper_plots/18_real_B.png',0)
img_V2 = cv.imread('paper_plots/60_real_A.png',0)
img_T2 = cv.imread('paper_plots/60_real_B.png',0)
my_fft_plots(img_V1, img_T1, img_V2, img_T2)
```
